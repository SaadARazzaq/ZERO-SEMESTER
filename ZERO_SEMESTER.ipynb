```python
# Task Day 3:

# Dataset used: https://www.kaggle.com/datasets/fedesoriano/airfoil-selfnoise-dataset

import pandas as pd

df = pd.read_csv('AirfoilSelfNoise.csv')

check_missing = df.isnull()  # Checks missing values in dataset

print('Are there any missing values in dataset: ', check_missing)

print()
print("--------------------------------------------------------------------")
print()

dup = df.drop_duplicates()

print("Duplicates dropped!")

print(dup)

print()
print("--------------------------------------------------------------------")
print()

import matplotlib.pyplot as plt

dup.hist(figsize=(16,8))
plt.show()
```

```python
# Task Day 4:

# Dataset used: https://www.kaggle.com/datasets/fedesoriano/airfoil-selfnoise-dataset

import pandas as pd

df = pd.read_csv('AirfoilSelfNoise.csv')

MEAN = df['alpha'].mean()

print("Mean is: ", MEAN)

MEDIAN = df['alpha'].median()

print("Median is: ", MEDIAN)

MODE = df['alpha'].mode()

print("Mode is: ", MODE)
```

```python
# Task Day 6:

# Dataset used: https://www.kaggle.com/datasets/fedesoriano/airfoil-selfnoise-dataset

import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv('AirfoilSelfNoise.csv')

X = df['alpha']
Y = df['delta']

plt.plot(X, Y)  #  Line Chart
plt.show()

plt.bar(X, Y)  #  Bar Chart
plt.show()
```

```python
# Task Day 7:

import numpy as np

my_array = np.array([2, 4, 6, 8, 10])

add = my_array + 5
sub = my_array - 5
mul = my_array * 5
div = my_array / 5

MEAN = np.mean(my_array)
STD = np.std(my_array)

print("Original array: ", my_array)
print("Addition: ", add)
print("Subtracttion: ", sub)
print("Multiplication: ", mul)
print("Division: ", div)
print("Mean: ", MEAN)
print("Standard deviation: ", STD)
```
